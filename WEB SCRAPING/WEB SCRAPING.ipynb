# Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame

# Write a python program to scrape product name, price and discounts from https://meesho.com/bags-ladies/pl/p7vbp

```python
# send request to webpage to get the source code
ProductPage = requests.get("https://meesho.com/bags-ladies/pl/p7vbp")
ProductPage

# getting page content
ProductSoup=BeautifulSoup(ProductPage.content)

#Fetching Product Names
ProductName=[] # define list variable
ProductNameSoup=ProductSoup.find_all('p',class_="bWSOET")
for ii in ProductNameSoup:
    ProductName.append(ii.text)

#Fetching Product Price
ProductPrice=[] # define list variable
ProductPriceSoup=ProductSoup.find_all('div',class_="eyya-Dr")
for ii in ProductPriceSoup:
    if int((ii.h5.text.replace('₹','')))>0:
        ProductPrice.append(ii.h5.text.replace('₹',''))

#print(ProductPrice)
    
#Fetching Product Discount
ProductDiscount=[] # define list variable
ProductDisSoup=ProductSoup.find_all('div',class_="eyya-Dr")
for ii in ProductDisSoup:
    ProductDiscount.append(ii.span.text.replace(" off",""))

#print(ProductDiscount)

MeeshoProducts=pd.DataFrame({'Product Name':ProductName,'Price':ProductPrice,'Discount':ProductDiscount})
MeeshoProducts
```

# Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.

```python
page=requests.get("https://coreyms.com/")
soup=BeautifulSoup(page.content)
data=soup.find_all("article",class_="post")
HData=[]
HDate=[]
HContent=[]
HUrl=[]
for i in data:
    HData.append(i.h2.text)
    HDate.append(i.time.text)
    HContent.append(i.find("div",class_="entry-content").p.text)
    if i.find("iframe",class_="youtube-player"):
        HUrl.append(i.find("iframe",class_="youtube-player")['src'])
    else:
        HUrl.append("No URL")
    #print(i.find("iframe",class_="youtube-player")['src'])
    

    
dFrame=pd.DataFrame({"Header":HData,"Date":HDate,"Content":HContent,"Video URL":HUrl})
dFrame
```

# Write a python program to scrape mentioned details from dineout.co.in : i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL 

```python
page=requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
soup=BeautifulSoup(page.content)
data=soup.find_all("div",class_="restnt-card restaurant")
RName=[]
RCuisine=[]
RLocation=[]
RRatings=[]
RImage=[]
for i in data:
    RName.append(i.find("a",class_="restnt-name ellipsis").text)
    RCuisine.append(i.find("div",class_="detail-info").text.split('|')[1])
    RLocation.append(i.find("div",class_="restnt-loc ellipsis").text)
    RRatings.append(i.find("div",class_="restnt-rating rating-4").text)
    RImage.append(i.find("img",{"data-gatype":"RestaurantImageClick"})['data-src'])
  
    
dFrame=pd.DataFrame({"Restaurant name":RName,"Cuisine":RCuisine,"Location":RLocation,"Ratings":RRatings,"Image URL":RImage})
dFrame
```
